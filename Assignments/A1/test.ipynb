```python
import numpy as np
import time

def monte_carlo_option_pricing(S0, K, r, T, sigma, N, num_paths):
    """
    Prices a European Call Option using Monte Carlo Simulation (Euler Scheme).
    
    Parameters:
    S0 (float): Initial stock price
    K (float): Strike price
    r (float): Risk-free interest rate
    T (float): Time to maturity (in years)
    sigma (float): Volatility
    N (int): Number of time steps
    num_paths (int): Number of simulated paths
    
    Returns:
    price (float): Estimated option price
    computation_time (float): Time taken for the simulation
    """
    start_time = time.time()
    dt = T / N  # Length of each time step
    
    # Initialize the stock price array for all paths
    # We simulate all paths simultaneously using NumPy vectorization for speed
    S = np.full(num_paths, S0)
    
    # Loop through each time step
    for _ in range(N):
        # Generate random standard normal variables (epsilon)
        z = np.random.normal(0, 1, num_paths)
        
        # Calculate the change in stock price (Delta S) using Euler discretization
        # Formula: Delta S = r * S * dt + sigma * S * epsilon * sqrt(dt)
        dS = S * (r * dt + sigma * z * np.sqrt(dt))
        
        # Update stock price
        S = S + dS
    
    # Calculate the payoff for a European Call Option: max(S_T - K, 0)
    payoffs = np.maximum(S - K, 0)
    
    # Calculate the average payoff and discount it back to present value
    # Price = exp(-rT) * mean(Payoffs)
    option_price = np.exp(-r * T) * np.mean(payoffs)
    
    end_time = time.time()
    computation_time = end_time - start_time
    
    return option_price, computation_time

# --- Execution Parameters ---
S = 7.92       # Spot Price
K = 7.92       # Strike Price
r = 0.0314     # Risk-free Rate (3.14%)
T = 0.5014     # Maturity (Years)
sigma = 0.202  # Volatility (20.2%)
steps = 120    # N time steps

# --- Run Simulations ---
# (a) 10,000 Paths
price_a, time_a = monte_carlo_option_pricing(S, K, r, T, sigma, steps, 10000)
print(f"Simulation (a) 10,000 paths: Price = {price_a:.5f}, Time = {time_a:.5f} sec")

# (b) 500,000 Paths
price_b, time_b = monte_carlo_option_pricing(S, K, r, T, sigma, steps, 500000)
print(f"Simulation (b) 500,000 paths: Price = {price_b:.5f}, Time = {time_b:.5f} sec")
```
